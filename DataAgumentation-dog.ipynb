```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
datagen = ImageDataGenerator(rotation_range=60,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,zoom_range=0.2,horizontal_flip=True,fill_mode='nearest')
```

```python
img = load_img(r'D:\DataScienceAndAICourse\May-NLP\Deep-learning\10th-may\DATA AUGMENTATION\dog.jpg')
```

```python
img
```

```python
x= img_to_array(img)
x= x.reshape((1,)+x.shape)

i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir=r'D:\DataScienceAndAICourse\May-NLP\Deep-learning\10th-may\DATA AUGMENTATION\images'):
     i+=1
     if i > 30:
        break
        
```
